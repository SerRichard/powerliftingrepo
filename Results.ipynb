```python
#RegressionAnalysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('regres.csv')
```

```python
data
```

```python
squat = pd.DataFrame(data[data['Lift'] == 'Squat'])
bench =  pd.DataFrame(data[data['Lift'] == 'Bench'])
deadlift = pd.DataFrame(data[data['Lift'] == 'Deadlift'])

fig = plt.figure(figsize=(7.5,7.5))
plt.suptitle('Accuracy change given knowledge', fontsize=16)
plt.xlabel('Knowledge of Athlete', fontsize=12)
plt.ylabel('R2 Score', fontsize=12)
plt.bar(squat['Knowledge'], squat['R2'], label='Squat R2', alpha=0.6, width=-0.25, align='edge')
plt.bar(bench['Knowledge'], bench['R2'], label='Bench R2', alpha=1, width=0.25)
plt.bar(deadlift['Knowledge'], deadlift['R2'], label='Deadlift R2', alpha=0.6, width=+0.25, align='edge')
plt.grid(True, color='purple', alpha=0.7, linestyle=':')
plt.legend()
fig.savefig('r2results.jpg')
```

```python
squat = pd.DataFrame(data[data['Lift'] == 'Squat'])
bench =  pd.DataFrame(data[data['Lift'] == 'Bench'])
deadlift = pd.DataFrame(data[data['Lift'] == 'Deadlift'])

fig = plt.figure(figsize=(7.5,7.5))
plt.suptitle('Error change given knowledge', fontsize=16)
plt.xlabel('Knowledge of Athlete', fontsize=12)
plt.ylabel('RMSE Score', fontsize=12)
plt.bar(squat['Knowledge'], squat['RMSE'], label='Squat RMSE', alpha=0.6, width=-0.25, align='edge')
plt.bar(bench['Knowledge'], bench['RMSE'], label='Bench RMSE', alpha=1, width=0.25)
plt.bar(deadlift['Knowledge'], deadlift['RMSE'], label='Deadlift RMSE', alpha=0.6, width=0.25, align='edge')
plt.grid(True, color='purple', alpha=0.7, linestyle=':')
plt.legend()
fig.savefig('RMSEresults.jpg')
```
